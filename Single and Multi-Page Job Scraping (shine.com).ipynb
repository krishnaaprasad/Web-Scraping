### Singlepage Job Web Scrapping

```python
from bs4 import BeautifulSoup
import requests
from csv import writer

position = str(input("Enter the Job Position: "))
city = str(input("Enter the City name for the Job: "))

url = f"https://www.shine.com/job-search/{position}-jobs-in-{city}-1?q={position},%20&loc={city}"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36",
    "Accept-Encoding": "gzip, deflate",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "DNT": "1",
    "Connection": "close",
    "Upgrade-Insecure-Requests": "1",
}

page = requests.get(url, headers=headers)
print(page)

soup_ = BeautifulSoup(page.content, 'html.parser')
soup= BeautifulSoup(soup_.prettify(), "html.parser")

containers = soup.findAll("div",{"class":"false"})

with open('Jobdetails.csv', 'w', encoding='utf8', newline='') as j:
    thewriter = writer(j)
    header = ['JobTitle','Company','Location','PostDate','JobUrl']
    thewriter.writerow(header)
    
    for container in containers:
        titles = container.find_all("h2",{'itemprop': 'name'})
        for title in titles:
            title_header = title.text.strip()
            print(title_header)
            companies= container.find_all("div",class_="jobCard_jobCard_cName__mYnow")
            for company in companies:
                companies_header = company.text.strip()
                print(companies_header)
                locations = container.find_all('div',{ 'class': 'jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2'})
                for location in locations:   
                    location_header = location.contents[0].strip()
                    print(location_header)
                    postdates = container.find_all('div', class_='jobCard_jobCard_features__wJid6')
                    for postdate in postdates:
                        postdate_header= postdate.text.strip()
                        print(postdate_header)
                        urls = container.find("h2",{'itemprop':'name'}).a
                        urls1=urls['href']
                        for url in urls1:
                            url_header="https://www.shine.com/"+ urls1
                        print(url_header)
                        detail=[title_header,companies_header,location_header,postdate_header,url_header]
                        thewriter.writerow(detail)
                    
                
```

### Multipage Job Web Scrapping

```python
from bs4 import BeautifulSoup
import requests
from csv import writer
import pandas as pd

Jobdetail_multipages = []

position = str(input("Enter the Job Position: "))
city = str(input("Enter the City name for the Job: "))

for i in range(1,6):
    url= f"https://www.shine.com/job-search/{position}-jobs-in-{city}-{i}?q={position},%20&loc={city}"
    headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36",
    "Accept-Encoding": "gzip, deflate",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "DNT": "1",
    "Connection": "close",
    "Upgrade-Insecure-Requests": "1"}
    
    page = requests.get(url, headers=headers)
    
    soup_ = BeautifulSoup(page.content, 'html.parser')
    soup= BeautifulSoup(soup_.prettify(), "html.parser")
    
    containers = soup.findAll("div",{"class":"false"})
    
    for container in containers:
        titles = container.find_all("h2", {'itemprop': 'name'})
        for title in titles:
            title_header = title.text.strip()
            

        companies = container.find_all("div", class_="jobCard_jobCard_cName__mYnow")
        for company in companies:
            companies_header = company.text.strip()
            

        locations = container.find_all('div', {'class': 'jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2'})
        for location in locations:
            location_header = location.contents[0].strip()
            

        postdates = container.find_all('div', class_='jobCard_jobCard_features__wJid6')
        for postdate in postdates:
            postdate_header = postdate.text.strip()
            

        urls = container.find("h2", {'itemprop': 'name'}).a
        urls1 = urls['href']
        for url in urls1:
            url_header = "https://www.shine.com/" + urls1
        

        Jobdetail_multipages.append([title_header, companies_header, location_header, postdate_header, url_header])

df = pd.DataFrame(Jobdetail_multipages, columns=['JobTitle','Company','Location','PostDate','JobUrl'])
print(df)
df.to_csv('Jobdetail_multipages.csv')        
    
    
```
